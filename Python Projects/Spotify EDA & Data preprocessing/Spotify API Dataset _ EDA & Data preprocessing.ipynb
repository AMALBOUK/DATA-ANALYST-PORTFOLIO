A partir du heatmap précédent, nous relevons les remarques suivantes :
- Une forte corrélation positive entre 'Loudness' et 'Energy' (0.7)
- Une corrélation positive observée entre 'Valence' d'un côté et 'Danceability', 'Energy', 'Loudness' d'un autre côté (0.4 à 0.5)
- Une corrélation négative remarquable entre 'Acousticness' et 'Energy' (-0.6)
- Une légère corrélation négative entre 'Acousticness' d'un côté et 'Loudness', 'Danceability' d'un autre côté (-0.4 à -0.3)

Nous pourrions conclure des deux pécédentes visualisations que les features les plus déterminants d'un succès (popularité), sont : Valence, Energy, Loudness, Danceability et Acousticness
