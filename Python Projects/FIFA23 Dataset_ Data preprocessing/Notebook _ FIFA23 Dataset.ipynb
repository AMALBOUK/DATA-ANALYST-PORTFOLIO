# 5- Encodage des données catégorielles

Nous passons maintenant au traitement des valeurs catégorielles.
Prenons la variable player_position, constituée de listes de valeurs séparées par des virgules, que nous allons transformer en one-hot encoding, après en avoir composé des groupes de positions pour réduire le nombre de features

Pour la variable player_positions, nous décidons tout d'abord de grouper ses valeurs pour ne pas monter inutilement en dimentionnalité.

```python
def group_positions(value):
    positions = {}
    positions['attacker'] = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW']
    positions['midfielder'] = ['LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LDM', 'CDM', 'RDM']
    positions['defender'] = ['LWB', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
    positions['keeper'] = ['GK']
    values = value.replace(' ','').split(',')
    for v in values:
        for position, val in positions.items():
            if v.upper() in val:
                return (position)
df['position'] = df['player_positions'].apply(lambda x: group_positions(x))
```

Pour examiner cette transformation, nous affichons les lignes des joueurs de l'équipe nationale marocaine.

```python
national_team = df[(df['nationality_id'] == 129) & (df['in_nation_team'] == 1)]
national_team.iloc[:, list(range(6)) + [-1]].head(20)
```

Nous continuons notre transformation pour arriver à quatre variables-groupe codées en one-hot encoding

```python
df_position = pd.get_dummies(df['position'])
df = pd.concat([df, df_position], axis=1)
df.drop(['player_positions', 'position'], axis=1, inplace=True)

national_team = df[(df['nationality_id'] == 129) & (df['in_nation_team'] == 1)]
national_team.iloc[:, list(range(6)) + [-1] + [-2] + [-3] + [-4]].head(20)
```

En poursuivant dans l'exploration et le prétraitement des données, nous allons convertir le format du taux d'activité des joueurs 'work_rate' en valeurs numériques dans deux nouvelles colonnes 'work_rate1' et 'work_rate2'. Ceci, afin d'obtenir une meilleure compréhension des données et de pouvoir exploiter ces variables dans les algorithmes de machine Learning.

```python
work_rate_dic = {}
work_rate_dic['Low'] = 1
work_rate_dic['Medium'] = 2
work_rate_dic['High'] = 3

def work_rate_v(value):
    value_list= value.split('/')
    w = [0, 0]
    for v in value_list:
        if w[0] == 0:
            w[0] = work_rate_dic[v]
        else:
            w[1] = work_rate_dic[v]
    return w
df['work_rate1'], df['work_rate2'] = zip(*df['work_rate'].apply(lambda x: work_rate_v(x)))
df.drop(['work_rate'], axis=1, inplace=True)
df
```
